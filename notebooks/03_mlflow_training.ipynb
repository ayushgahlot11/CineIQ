```python
import mlflow
import mlflow.sklearn
from surprise import Reader, Dataset, SVD
import pandas as pd
import pickle
import os

print("Setting up MLflow tracking...")

# 1. Setup MLflow Experiment
mlflow.set_experiment("CineIQ_Collaborative_Filtering")

# Load data
processed_dir = "../data/processed/"
ratings = pd.read_csv(os.path.join(processed_dir, 'ratings_sample.csv'))
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# 2. Define our hyperparameters
params = {"n_epochs": 20, "lr_all": 0.005, "reg_all": 0.02}

# 3. Train and Log the run
with mlflow.start_run():
    mlflow.log_params(params)
    
    svd_model = SVD(**params)
    trainset = data.build_full_trainset()
    svd_model.fit(trainset)
    
    # Save the model locally and log it to MLflow
    os.makedirs("../artifacts/saved_models/", exist_ok=True)
    model_path = "../artifacts/saved_models/svd_model_tracked.pkl"
    with open(model_path, "wb") as f:
        pickle.dump(svd_model, f)
        
    mlflow.log_artifact(model_path)
    print("Success! SVD Model trained and logged to MLflow.")
```
